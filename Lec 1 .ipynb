### Arth. Operations

### Swapping Values of Variables

```python
m = int(input("Enter the Number a: "))
n = int(input("Enter the Number b: "))

#Method One (Addition Subraction Method)
m = m + n
n = m - n
m = m- n
print("Method 1")
print("M: ",m)
print("N: ",n)

#Method Two (Temp Variable)
t = m
m = n
n = t
print("Method 2")
print("M: ",m)
print("N: ",n)

#Method Three (, Operator)
m,n = n,m
print("Method 3")
print("M: ",m)
print("N: ",n)
```

##### Use of % in Printing

```python
#Out Printed will have total digit mentioned in the print statement
#example: %06d will have 6 digits in total 000002
x = 2
print("%06d"%x)
print("%05d"%x)
print("%04d"%x)
print("%02d"%x)
print("%01d"%x)
```

```python
#Same as above but negative
x = 2
print("-%06d"%x)
print("-%05d"%x)
print("-%04d"%x)
print("-%02d"%x)
print("-%01d"%x)
```

```python
#Subraction sign after Percent/Mod Sign Left Justified and the number
#defines the number of spaces after the justified number
x = 2
print("%-06d"%x)
print("%-05d"%x)
print("%-04d"%x)
print("%-02d"%x)
print("%-01d"%x)
```

```python
#f number mentions the Number of Decimal Points
x = 56
print("%.2f"%x)
print("%.3f"%x)
print("%.4f"%x)
print("%.5f"%x)
```

```python
x = 3.3256789
print(' %-6.2f | %05.2f | %+06.1f' %(x, x, x))
```

```python
name = "abcdefgh"
pie = 3.14159
print("%s"%name)
print("%s"%pie)
print("%d"%pie)
print("%f"%pie)
print("%.3f"%pie)
print("%g"%pie)
```
