###  1.) Python Program to accept four numbers and print the Average of Numbers

```python
a = int(input("Enter number a: "))
b = int(input("Enter number b: "))
c = int(input("Enter number c: "))
d = int(input("Enter number d: "))

avg = (a + b + c + d) / 4
print("Average of 4 Numbers is: ",avg)
```

### 2.)	Python Program to Exchange the Values of Two Numbers Without Using a Temporary Variable

```python
x = int(input("Enter 1st Number: "))
y = int(input("Enter 2nd Number: "))

#Swapping Method 1
x,y = y,x
print("New Values after Swapping Method 1")
print("x: ",x)
print("y: ",y)

#Swapping Method 2
x += y
y = x - y
x -= y
print("New Values after Swapping Method 2")
print("x: ",x)
print("y: ",y)
```

### 3.)	Python Program to Read a number n and Compute n+nn+nnn using for loop

```python
n = int(input("Enter the Value of n: "))
expo = int(input("Enter the final term: "))
ans = 0
for i in range(1,expo + 1):
    ans += n ** i
    
print("Result: ",ans)
```

### 4.)	Python Program to Reverse a Given Number

```python
j = int(input("Enter the Value of j: "))
rev = 0
while j > 0:
    rev *= 10
    rev += (j % 10)
    j = j//  10

print("Reversed Number: ",rev)
```

### 5.)	Python Program to Check Whether a Number is Positive or Negative

### 6.)	Python Program to Take in the Marks of 5 Subjects calculate avg and Display the Grade

```python
s1 = int(input("Enter number subject1: "))
s2 = int(input("Enter number subject2: "))
s3 = int(input("Enter number subject3: "))
s4 = int(input("Enter number subject4: "))
s5 = int(input("Enter number subject5: "))

avgmks = (s1+s2+s3+s4+s5)/5

print("Average Marks: ",avgmks)
print("Grade: ")
if avgmks>80:
    print("A")
elif avgmks>60:
    print("B")
elif avgmks>40:
    print("C")
else:
    print("F")
```

### 7.)	Python Program to Print all Numbers in a Range Divisible by a Given Number

```python
r = int(input("Enter a Number: "))
rang = int(input("Enter the Range: "))

print("Number Divisble in the Given Range:")
for t in range(1,rang):
    if t % r == 0:
        print(t)
```

### 8.)	Python Program to Read Two Numbers and Print Their Quotient and Remainder

### 9.)	Python Program to Print Odd Numbers Within a Given Range

```python
ra = int(input("Enter Range: "))
print("Odd Numbers in Range: ")
for k in range(1,ra+1):
    if k % 2:
        print(k)
```

### 10.)	Python Program print factorial of a given number 

```python
number = int(input("Enter the Number for Factorial: "))
facts = number

for i in range(1,number):
    facts *= i

print("Factorial is: ",facts)
```
