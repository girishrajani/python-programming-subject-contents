### Q1 Python Program to Find the Sum of Digits in a Number

```python
num = int(input("Enter the Number: "))
sum = 0
while num != 0:
    sum += (num%10)
    num //= 10

print("Sum of Digits: ",sum)
```

### Q2 Python Program to Find the Smallest Divisor of an Integer

```python
n = int(input("Enter the Number: "))

for i in range(2,n):
    if n%i==0:
        smallestDivisor = i
        break
print("Smallest Divisor is ",smallestDivisor)
```

### Q3 Python Program to Count the Number of Digits in a Number

```python
num = int(input("Enter the Number: "))
count = 0
while num != 0:
    count += 1
    num //= 10

print("Number of Digits: ",count)
```

### Q4 Python Program to Check if a Number is a Palindrome

```python
j = int(input("Enter the Value of j: "))
num = j
rev = 0
while j > 0:
    rev *= 10
    rev += (j % 10)
    j = j//  10

if num == rev:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

### Q7 Python Program to Read a Number n and Print the Natural Numbers Summation Pattern

```python
k = int(input("Enter a Number: "))
for i in range(1,k+1):
    sum = 0
    for j in range(1,i+1):
        sum += j
        print(j,end='')
        if(j<i):
            print("+",end='')
    
    print(" = ",sum)
```

### Q8 Python Program to Print an factorial of a number

```python
number = int(input("Enter the Number for Factorial: "))
facts = number

for i in range(1,number):
    facts *= i

print("Factorial is: ",facts)
```
